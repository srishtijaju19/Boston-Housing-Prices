```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

### Importing Data set

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

### Exploratory analysis

```python
# to find count of rows and columns in the data
print(boston.data.shape)
```

```python
#print column names
print(boston.feature_names)
```

```python
#print description of the column names
print(boston.DESCR)
```

```python
#converting to pandas
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
# converting index to column names
bos.columns = boston.feature_names
bos.head()
```

```python
# shape of the target variable PRICE which is not in the data frame
print(boston.target.shape)
```

```python
#adding PRICE to the data frame
bos['Price'] = boston.target
bos.head()
```

```python
#summary of the data
bos.describe()
```

### Splitting the dataset

```python
#X are the predictor values and Y is the target value
X = bos.drop('Price', axis = 1)
Y = bos['Price']
```

```python
#splitting the data set
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 5)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, Y_train)

Y_pred = lm.predict(X_test)
```

```python
#plot the predicted values
area = np.pi*3

plt.scatter(Y_test, Y_pred, alpha = 0.5, s = area)
plt.title("Prices vs Predicted Prices")
plt.xlabel("Prices")
plt.ylabel("Predicted Prices")
```

### Mean Squared Error

```python
# mse indicates the level of error of a model. MSE will check the difference between the actual value and predicted value
mse = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
print(mse)
```
